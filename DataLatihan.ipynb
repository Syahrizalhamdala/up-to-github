```python
import numpy as np
import pandas as pd

n_data = 100

usia = np.random.randint(18,61,size=n_data)
transaksi = np.random.randint(0,51,size=n_data)
label = np.random.randint(0,2,size=n_data)
gander = np.random.choice(['L','P'],size=n_data)

df = pd.DataFrame({
    'Usia' : usia,
    'Transaksi' : transaksi,
    'label' : label,
    'Gender' : gander
    
})
df.to_csv('dataLatih.csv',index=False)
df.head #mengetahui 5 baris pertama dari dataframe
df.info()
```

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

np.random.seed(42)

n = 10000

# Buat list tanggal acak langsung tanpa fungsi
start = datetime(2020, 1, 1)
end = datetime(2025, 1, 1)
start_u = start.timestamp()
end_u = end.timestamp()
timestamps = [datetime.fromtimestamp(random.uniform(start_u, end_u)) for _ in range(n)]

# Buat DataFrame simulasi data Bitcoin
df = pd.DataFrame({
    'timestamp': timestamps,
    'price_usd': np.round(np.random.uniform(10000, 70000, n), 2),
    'volume_btc': np.round(np.random.uniform(0.01, 100.0, n), 4),
    'market_cap_usd': np.round(np.random.uniform(1e9, 1.5e12, n), 2),
    'transactions': np.random.randint(50, 10000, n),
    'mining_difficulty': np.round(np.random.uniform(1e12, 3e13, n), 2),
    'hash_rate_ths': np.round(np.random.uniform(50000, 250000, n), 2),
    'fees_usd': np.round(np.random.uniform(0.1, 500.0, n), 2),
    'sentiment': np.random.choice(['Bullish', 'Bearish', 'Neutral'], n),
    'exchange': np.random.choice(['Indodax', 'Triv', 'Pintu', 'Pluang'], n)
})

# Urutkan berdasarkan waktu
df = df.sort_values('timestamp').reset_index(drop=True)

# Lihat 5 data pertama
df.head()

```
