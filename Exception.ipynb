```python

try:
    angka1 = int(input('Angka 1:'))
    angka2 = int(input('Angka 2:'))
    operator = input('Operator (+,-,*,/):')
    hasil = 0
    if operator == '+':
        hasil = angka1 + angka2
    elif operator == '-':
        hasil = angka1 - angka2
    elif operator == '*':
        hasil = angka1 * angka2
    elif operator == '/':
        hasil = angka1 / angka2
    else:
        hasil = 'operator tidak ditemukan'

except (ZeroDivisionError, ValueError) as arr:
    print("terdapat kesalahan:", arr)
else:
    print(f"Hasil : {hasil}")
finally:
    print("proses perhitungan selesai di kerjakan")
```

```python
#IndexError -> LIST
data = []
try:
    jumlahData = int(input('banyak data:'))

    for i in range (0,jumlahData):
        nilai = int(input('isikan nilai siswa:'))
        data.append(nilai)


    showData = int(input('Tampilkan data urutan ke berapa: '))
    hasil = data[showData-1]
except ValueError:
    print("input harus angka")
except IndexError:
    print("data tidak tersedia")
else:
    print(hasil)
finally:
    print('selamat kamu berhasil')
```

```python
#Custom Exception
#Tiket Film, umur harus >= 17

class UmurException(Exception):
    pass

def checkUmur(umur):
    if umur < 17:
        raise UmurException("Umur harus lebih dari 16")

try:
    umur = int(input("isikan umur: "))
    checkUmur(umur)
except (ValueError, UmurException) as arr:
    print("terjadi kesalahan:", arr)
else:
    print("umur anda valid")
finally:
    print("cek umur selesai")
```
